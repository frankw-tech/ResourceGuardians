```python
import matplotlib.pyplot as plt

usage_types = ["Teaching", "Resource", "Dorm"]
average_costs = [1250, 2000, 800]

plt.bar(usage_types, average_costs, color='blue')

plt.xlabel("Type of Usage")
plt.ylabel("Average Cost of Electricity")
plt.title("Average Cost of Electricity per Type of Usage")

plt.show()
```
