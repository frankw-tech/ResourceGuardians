**Load bike-sales CSV File into df dataframe and display the first 5 rows**

```python
import pandas as pd
df = pd.read_csv('Sample.csv')
df.head()
```

**Take a deep copy of df and save it to df1**

```python
df1=df.copy()
```

**Import required libraries for visualization (matplotlib and seaborn)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1['Electricity per SF'] = df1['Electricity'] / df1['Square Footage'] 
plt.figure(figsize=(10, 6))  # Set the figure size (optional) 
sns.barplot(x=df1['Building Name'], y=df1['Electricity per SF'])
plt.title('Electricity Cost per Square Footage for Each Building')
plt.xlabel('Building Name') 
plt.ylabel('Electricity per SF (Cost/SF)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
